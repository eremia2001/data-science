__Teilaufgabe d)__ Stellen Sie die absoluten Häufigkeiten und die empirische Wahrscheinlichkeitsfunktion auf Basis der gespeicherten Ergebnisse für jeden möglichen Würfelwert aus der vorherigen Teilaufgabe graphisch dar.

```python
import numpy as np
import matplotlib.pyplot as plt


def erstelle_diagramm(werte,haeufigkeiten,ergebnis):
    # Schritt 2: Berechnung der empirischen Wahrscheinlichkeitsfunktion (relative Häufigkeit)
    empirische_wahrscheinlichkeiten  = haeufigkeiten / len(ergebnis) # Nur möglich so zu teilen ,weil hauefigkeit Array ein np.ndarray vom Typ ist 
    print(f'Emprische Warhscheinlichkeit {empirische_wahrscheinlichkeiten}' )

    # Schritt 3: Grafische Darstellung
    fig, axs = plt.subplots(2, 1, figsize=(8, 10))

    axs: list[plt.Axes]  # Typ-Hinweis für die Variable damit VsCode das erkennt 

    # Balkendiagramm für die absoluten Häufigkeiten
    axs[0].bar(werte,haeufigkeiten,color="skyblue")
    axs[0].set_title('Absolute Häufigkeit der Würfelwürfe')
    axs[0].set_xlabel('Würfelwert')
    axs[0].set_ylabel('Absolute Häufigkeit')

    # Balkendiagramm für die empirische Wahrscheinlichkeitsfunktion
    axs[1].bar(werte,empirische_wahrscheinlichkeiten)
    axs[1].set_title('Empirische Wahrscheinlichkei der Würfelwürfe')
    axs[1].set_xlabel('Würfelwert')
    axs[1].set_ylabel('Empirische Wahrscheinlichkeit')

    # Zeige das Diagramm an
    plt.tight_layout()
    plt.show()
    
# Schritt 1: Berechnung der absoluten Häufigkeiten
werte , hauefigkeit = np.unique(wuerfel_wuerfe,return_counts=True)
print(f'Werte der einzelnen Würfel Würfe {werte} die Absoluten Häufigkeiten der Werte in der selben Reihenfolgen {hauefigkeit}' )
print(f'Typen des Array mit den Werten {type(werte)} Typen des Array mit Häufigkeiten {type(hauefigkeit)}' )
erstelle_diagramm(werte,hauefigkeit,wuerfel_wuerfe)
```
